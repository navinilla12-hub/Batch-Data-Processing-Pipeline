```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
```

```python
# Step 2: Load Dataset (Batch Ingestion)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/flights.csv"
df = pd.read_csv(url)

print("Raw Data:")
print(df.head())
```

```python
# Step 3: Data Cleaning
df.columns = df.columns.str.lower().str.strip()
```

```python
# Handle missing values (if any)
df = df.dropna()
```

```python
# Convert types
df['passengers'] = df['passengers'].astype(int)
```

```python
# Step 4: Data Transformation
# Create new features
df['year_month'] = df['year'].astype(str) + "-" + df['month']
```

```python
# Aggregate data
monthly_summary = df.groupby('month')['passengers'].mean().reset_index()
yearly_summary = df.groupby('year')['passengers'].sum().reset_index()
```

```python
# Sort data
df = df.sort_values(by=['year', 'month'])
```

```python
# Step 5: Data Validation
def validate_data(data):
    assert data.isnull().sum().sum() == 0, "Missing values found!"
    assert (data['passengers'] >= 0).all(), "Negative values found!"
    print("Data Validation Passed ✅")

validate_data(df)
```

```python
# Step 6: Save Processed Data (Batch Output)
df.to_csv("processed_flights_data.csv", index=False)
monthly_summary.to_csv("monthly_summary.csv", index=False)
yearly_summary.to_csv("yearly_summary.csv", index=False)
```

```python
# Step 7: Logging Output
print("\nPipeline Execution Completed Successfully 🚀")
print("Processed files saved:")
print("- processed_flights_data.csv")
print("- monthly_summary.csv")
print("- yearly_summary.csv")
```
